# Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

# importing Boston House Price Dataset

```python
house_data = sklearn.datasets.load_boston()
```

```python
house_data
```

```python
house_df = pd.DataFrame(house_data.data, columns = house_data.feature_names)
```

```python
house_df
```

```python
house_df['Price'] = house_data.target
```

```python
house_df.head()
```

```python
house_df.shape
```

```python
house_df.describe()
```

```python
house_df.isnull().sum()
```

# Understanding Correlation 

```python
correlation = house_df.corr()
```

```python
# Heatmap for understanding the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

# Train Test Split

```python
x = house_df.drop(columns = 'Price', axis = 1)
y = house_df['Price']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Model Training
XGBoost Regressor

```python
model = XGBRegressor()
```

```python
model.fit(x_train, y_train)
```

# Model Evaluation

```python
# Accuracy on the train data
x_train_pred = model.predict(x_train)
# R-squared error
r2_score = metrics.r2_score(y_train, x_train_pred)
print(r2_score)
# Mean absolute Error
mse_score = metrics.mean_absolute_error(y_train, x_train_pred)
print(mse_score)
```

```python
# Accuracy on test data
x_test_pred = model.predict(x_test)
# R-squared error
r2_score = metrics.r2_score(y_test, x_test_pred)
print(r2_score)
# Mean absolute Error
mse_score = metrics.mean_absolute_error(y_test, x_test_pred)
print(mse_score)
```

```python
# Visualize the actual prices and predicted prices
plt.scatter(y_train, x_train_pred)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual VS predicted')
plt.show()
```
